```python
#compare monolix model output (likelihood and AIC etc)

mnames = [
    '0carrycap',
    '0otherlogistic',
    '1nodiff',
    '2nodiff-mustclear',
    '3diff',   
    '3diff-paramconstrained',   
    '4diff_form2',
    '6diff-corrmat',#'6diff-corrmat_backdiff','6diff-corrmat_1TEMoutrate',
    '6diff-corrmat_paramconstrained',
    '6diff-corrmat_paramconstrained-logit',
    '10alldiff',
]

dfl=[]
for mn in mnames:
    df = pd.read_csv('../monolix2/'+mn+'/LogLikelihood/logLikelihood.txt')
    
    dfl.append([mn,df.iloc[0,1],df.iloc[1,1],df.iloc[2,1],df.iloc[3,1]])
    
comp_df = pd.DataFrame(dfl,columns=['model_name','-2LL','AIC','BIC','BICc'])

sorted_df = comp_df.sort_index(by='AIC').round(decimals=1)

Nbackcalc=(sorted_df['AIC']-sorted_df['-2LL'])/2 #use AIC to back calculate the number of parameters estimated
#sorted_df['total N']=(sorted_df['AIC']-sorted_df['-2LL'])/2 #use AIC to back calculate the number of parameters estimated
sorted_df['N rates']=(Nbackcalc-10-5-3)/2 #-10 for T0 and omegaT0, -5 for b noise, -3 for corrs
sorted_df['ΔLL']=sorted_df['-2LL']-sorted_df['-2LL'].iloc[0]
sorted_df['ΔAIC']=sorted_df['AIC']-sorted_df['AIC'].iloc[0]
sorted_df['ΔBIC']=sorted_df['BIC']-sorted_df['BIC'].iloc[0]

#sorted_df[['model_name', '-2LL', 'AIC', 'N rates', 'ΔAIC', 'BICc']]

outdf=sorted_df[['model_name', 'ΔLL', 'N rates', 'ΔAIC', 'ΔBIC']]
outdf
```
